```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=3)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')#kind=hex ,scatter,reg(regression)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')#jointplot for all columns as much as possible
            #hue=categorical data(yes/no,female/male)not numerical
```

```python
sns.rugplot(tips['total_bill'])#draws dashmark for every points
```

```python
#categorical data
#bar graph
#data not part of exapmle datasets
#sns.barplot(x=df.sex,y=df.tota_bill,data=tips,estimator=np.std)
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)#count no of occurances(y axis)
```

```python
#boxplots
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')#violinplot+stripplot
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
tc=tips.corr()#correlation data
tc
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')#annot=display values in each cell
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(fp,cmap='magma',linecolor='white',linewidth=1)
```

```python
#cluster map
sns.clustermap(fp)
```

```python
#grids
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

```python
#regression plots
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='sex',row='time',markers=['o','v'],scatter_kws={'s':100})
```

```python
#style and color
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine(left=True)
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```
