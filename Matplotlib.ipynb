```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x=np.linspace(0,5,11)
y=x ** 2
plt.plot(x,y,'r-')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,y,'g')
```

```python
#OO
fig=plt.figure()
axes=fig.add_axes([1.0,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Graph')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for curraxes in axes:
    curraxes.plot(x,y)
axes[0].set_title('Graph 1')
axes[1].set_title('Graph 2')
#plt.tight_layout()
```

```python
#figure size and DPI
fig=plt.figure(figsize=(10,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='Grapgh1',color='green')
ax.plot(x,x**3,label='Graph2')
ax.legend(loc=1)

```

```python
fig.savefig('mypic.png',dpi=200)
```

```python
#plot appearance
fig=plt.figure()
ax=fig.add_axes([2,2,1,1])
ax.plot(x,y,color='black',linewidth=2,alpha=0.5,linestyle='-',marker='o',markersize=10,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green') #linewidth==lw, linestyle==ls
                                                                            # marker= marks the points
ax.set_xlim([0,1])
ax.set_ylim([0,1])
```
